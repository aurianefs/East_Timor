East Timor, also known as Timor-Leste, is a small country located in Southeast Asia. Currently, Timor-Leste is a young country,
having gained independence from Indonesia in 2002. The island of Timor was named by the Malays due to its location to the east. 
"Timur" in Malay means "east," so the island became known as Timor.

Timor-Leste has a complex history, shaped by centuries of colonization, conflict, and struggle for independence. During World 
War II, Timor-Leste was occupied by Japanese forces, who used it as a strategic base to launch attacks on other parts of 
Southeast Asia. The occupation of Timor-Leste was marked by acts of brutality and violence against the local population, with 
many Timorese losing their lives in the process. In the post-war period, Portugal returned to Timor-Leste and reestablished its
colonial rule.

Despite centuries of missionary colonization by Portugal, the country never gave Timor-Leste the same level of political 
importance as its other colonies in Southeast Asia. This was partly due to Timor-Leste's remote location and rugged terrain, 
which made access and control difficult. The Portuguese presence in Timor-Leste was primarily focused on economic exploitation,
namely the production of sandalwood and coffee. As a result of Portugal's neglect of Timor-Leste, the region remained 
economically underdeveloped and politically marginalized for much of its history.

By the mid-17th century, the island of Timor had become a contested zone between the Portuguese and Dutch colonial powers. The 
Portuguese established a presence in Timor in the early 16th century, primarily due to the trade of sandalwood and spices. As a
result, many Portuguese soldiers and civilians decided to leave Timor and return to Portugal or other colonies in Southeast Asia.
On February 20, 1942, Japan invaded Timor-Leste as part of its broader military campaign in Southeast Asia during World War II.

In 1945, as the tide of the war began to turn against Japan, Australian forces landed in Timor-Leste and engaged in a campaign 
to expel the remaining Japanese troops. Australian military administered Timor-Leste until Portugal regained control of the 
region in 1946. One of the consequences of the Carnation Revolution was the abandonment of Portugal's remaining colonies, 
including Timor-Leste. However, Indonesia, which had long coveted Timor-Leste for its natural resources and strategic location,
saw an opportunity to assert its dominance in the region.

Just nine days after Timor-Leste declared its independence, Indonesian forces invaded the country, initiating a brutal and 
prolonged conflict that would last for more than two decades. The Indonesian invasion of Timor-Leste was marked by acts of 
violence and brutality against the local population. Indonesian troops, commanded by General Benny Moerdani, invaded Timor-Leste on December 7, 1975, using a combination of land, air, and sea forces to subdue the small and poorly equipped Timorese defense forces. Finally, in 1999, after years of international pressure and growing unrest within Indonesia itself, the Indonesian government agreed to a referendum on Timorese independence.

Timor-Leste officially became an independent nation on May 20, 2002.

# Conclusion
The research reveals several significant findings. Among the main points, we can conclude that:

1. Approval of Teaching: The majority of Timorese students approved of the teaching at UNILAB, indicating the effectiveness of the teaching methodologies adopted by the institution.
2. Expectations versus Reality: Many students were surprised to discover that the city of Redenção did not match the images that were shown to them, which were actually of Fortaleza.
3. Language Barriers: The difficulty with Portuguese is due to the fact that Timorese students learned European Portuguese before traveling, not Brazilian Portuguese.
4. Motivation and Responsibility: Timorese students came to Brazil not only for the opportunity to study abroad but also with the responsibility of acquiring knowledge to develop Timor upon their return.
5. Limited Curiosity: The majority of students did not seek detailed information about UNILAB before their arrival, which contributed to their surprise upon encountering a reality different from what they had imagined.
6. Adaptation Challenges: The data revealed significant challenges faced by Timorese students at UNILAB, mainly related to language, learning, and cultural adaptation. These insights can serve as a guide for future research and initiatives aimed at improving the reception and productivity of international students.

These conclusions provide a comprehensive overview of the experience of Timorese students at UNILAB and the complexities of their adaptation. They not only highlight the effectiveness of the teaching methodologies and the students' motivations but also underscore the areas that need attention to facilitate better integration and academic success for these students in Brazil.
